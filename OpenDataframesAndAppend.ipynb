```python
import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
#Function that append each text file to the dataframe
def openAndAppend(df, path, drop_duplicates=False, column_to_drop=""):
    if drop_duplicates == True and column_to_drop == "":
        raise Exception("Wrong parameters")
    onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
    print("Total files in ",path,"folder:",len(onlyfiles))
    counter = 1
    for file in onlyfiles:     
        if file == ".DS_Store":
            continue
        else:
            #path
            filepath = str(path + "/"+ file)
            
            #Print count 
            print(filepath, "\n File number:",counter,"/",len(onlyfiles))
            print("")
            counter += 1
            
            #open the file, read and put it into a tuple with the author
            df2 = pd.read_csv(filepath)

            #Merge
            df = df.append(df2)
            if drop_duplicates == True:
                df = df.drop_duplicates(subset=[column_to_drop])
    return df
```

```python
df = openAndAppend(df,"/Users/gadgethub/backup/Indeed",drop_duplicates=True,column_to_drop="jobkeys")
```
