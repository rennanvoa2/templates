```python
import pandas as pd
import xlwt
from tempfile import TemporaryFile

df = pd.read_csv("/Users/gadgethub/Downloads/Esco Inglês/ISCOGroups_en.csv")
```

```python
df.loc[df["code"] == "011"]["preferredLabel"]
```

```python
df["code"] = df["code"].apply(str)
```

```python
#add 0 to the beggin of army codes
for i in range(1,10):
    df["code"][i] = "0"+ str(df["code"][i])
```

```python
jobs = df[["preferredLabel","code"]]
```

```python
parents = []
major = []
minor = []
brand = []

for i,v in enumerate(jobs["code"]):
    if len(str(v)) == 1:
        parents.append((v,jobs["preferredLabel"][i]))
    if len(str(v)) == 2:
        major.append((v,jobs["preferredLabel"][i]))
    if len(str(v)) == 3:
        minor.append((v,jobs["preferredLabel"][i]))
    if len(str(v)) == 4:
        brand.append((v,jobs["preferredLabel"][i]))
```

```python
parents
```

```python
book = xlwt.Workbook()
sheet1 = book.add_sheet('parents')
sheet2 = book.add_sheet('major')
sheet3 = book.add_sheet('minor')
sheet4 = book.add_sheet('brand')

for i,e in enumerate(parents):
    sheet1.write(i,0,e[0])
    sheet1.write(i,1,e[1])
    
for i,e in enumerate(major):
    sheet2.write(i,0,e[0])
    sheet2.write(i,1,e[1])
    
for i,e in enumerate(minor):
    sheet3.write(i,0,e[0])
    sheet3.write(i,1,e[1])

for i,e in enumerate(brand):
    sheet4.write(i,0,e[0])
    sheet4.write(i,1,e[1])
    
name = "/Users/gadgethub/Google Drive/Trabalho/Neobrain Projects/Esco_groups/hierarchy.xls"
book.save(name)
book.save(TemporaryFile())
```

```python
codes = []
groups = []
for ind,i in enumerate(parents):
    group = []
    code = []
    for index,value in enumerate(jobs["code"]):
        if str(value).startswith(str(i[0])):
            group.append(jobs["preferredLabel"][index])
            code.append(jobs["code"][index])
    codes.append(code)
    groups.append(group)    
```

```python
for i,e in enumerate(groups):
    df_temp = pd.DataFrame()
    df_temp["code"] = codes[i]
    df_temp["name"] = groups[i]
    print(df_temp)
```

```python
book = xlwt.Workbook()

counter = 0
for i,e in enumerate(groups):
    df_temp = pd.DataFrame()
    df_temp["code"] = codes[i]
    df_temp["name"] = groups[i]
    sheet1 = book.add_sheet(str(counter))
    for i,e in enumerate(df_temp["code"]):
        sheet1.write(i,0,e)
    for i,e in enumerate(df_temp["name"]):
        sheet1.write(i,1,e)
    counter += 1
name = "/Users/gadgethub/Google Drive/Trabalho/Neobrain Projects/Esco_groups/groups.xls"
book.save(name)
book.save(TemporaryFile())
```
